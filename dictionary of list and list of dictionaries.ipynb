list of dictionaries

create a list of 3 dictionaries

```python
students = [
    {"name": "Amit", "age": 20},
    {"name": "Neha", "age": 22},
    {"name": "Rahul", "age": 19}
]
```

name of first student

```python
print(students[0]["name"])
```

loop and print all names

```python
for student in students:
    print(student["name"])
```

Add new student Priya (age 21)

```python
students.append({"name": "Priya", "age": 21})
print(students)
```

Update Rahul’s age to 20

```python
for student in students:
    if student["name"] == "Rahul":
        student["age"] = 20
print(students)
```

Remove Neha from the list

```python
for s in students:
    if s["name"] == "Neha":
        students.remove(s)
print(students)
```

Find the oldest student

```python
oldest = students[0]

for s in students:
    if s["age"] > oldest["age"]:
        oldest = s

print(oldest)
```

Print students older than 20

```python
for student in students:
    if student["age"] > 20:
        print(student)
```

Convert to list of names only


```python
names = []

for s in students:
    names.append(s["name"])

print(names)
```

Sort students by age

```python
for i in range(len(students)):
    for j in range(i + 1, len(students)):
        if students[i]["age"] > students[j]["age"]:
            students[i], students[j] = students[j], students[i]

print(students)
```

dictionary of lists

initial data

```python
students = {
    "names": ["Amit", "Neha", "Rahul"],
    "ages": [20, 22, 19]
}
print(students)
```

Print first student’s name and age

```python
print(students["names"][0], students["ages"][0])
```

Add new student Priya (21)

```python
students["names"].append("Priya")
students["ages"].append(21)
print(students)
```

Update Rahul’s age to 20

```python
index = students["names"].index("Rahul")
students["ages"][index] = 20
print(students)
```

Remove Neha and her age

```python
index = students["names"].index("Neha")
students["names"].pop(index)
students["ages"].pop(index)
print(students)
```

Find student with maximum age

```python
max_age = max(students["ages"])
index = students["ages"].index(max_age)
print(students["names"][index], max_age)
```

Print students older than 20

```python
for i in range(len(students["names"])):
    if students["ages"][i] > 20:
        print(students["names"][i], students["ages"][i])
```

Convert to list of dictionaries

```python
list_of_dicts = []

for i in range(len(students["names"])):
    list_of_dicts.append({
        "name": students["names"][i],
        "age": students["ages"][i]
    })

print(list_of_dicts)
```

Convert to Pandas DataFrame

```python
import pandas as pd
```

```python
df = pd.DataFrame(students)
print(df)
```

Sort students by age

```python
sorted_students = sorted(
    zip(students["names"], students["ages"]),
    key=lambda x: x[1]
)

print(sorted_students)
```
