```python
def count_vowels(s):
    count = 0
    vowels = "aeiouAEIOU"
    for char in s:
        if char in vowels:
            count+=1
    return count
string = "elli"
print(count_vowels(string))
```

```python
def factoria(n):
    fact = 1
    for i in range(1,n+1):
        fact*=i
    return fact
n = 5
print(factoria(n))
```

```python
def is_palindrome(s):
    if is_palindrome(s):
       s==s[::-1]
    print("palindrome")
    else:
    print("not palindrome")
s="radar"
print(is_palindrome(s))
    
```

```python
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact(n-1)
n=5
print(fact(n))
```

```python
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)
n=7
print(fib(n))
```

```python
def topper(students):
    max_marks = 0
    top_name = ''
    for s in students:
        if s['marks']>max_marks:
            max_marks =s['marks']
            top_name = s['name']
        return top_name
students = [{'name':'alice','marks':85},{'name':'bob','marks':92},{'name':'charlie','marks':88}]
topper(students)            

            
    
```
