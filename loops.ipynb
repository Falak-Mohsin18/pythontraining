FACTORIAL NUMBER


```python
num = int(input("enter number:"))
f =  1
while num > 0:
    f = f * num
    num = num-1
print(f)
```

REVERSE A NUMBER

```python
n = 123
r = 0

while n > 0:
    r = r * 10 + n % 10
    n //= 10

print(r)
```

SUM OF DIGITS

```python
num = int(input("Enter a number: "))
sum = 0

while num > 0:
    sum += num % 10
    num //= 10

print("Sum of digits:", sum)
```
