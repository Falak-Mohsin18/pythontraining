datetime module

```python
from datetime import datetime,date,timedelta
```

```python
print(datetime.now())
```

```python
print(date.today())
```

```python
today = date.today()
print(today.year,today.month,today.day)
```

```python
now = datetime.now()
print(now.strftime("%A,%B,%d,%I:%M:%p"))
```

```python
dob = date(2005,2,17)
today = date.today()
age = today.year - dob.year
print(age)
```

```python
today = date.today()
end = date(today.year,12,31)
print((end - today).days)
```

```python
d1 = date(2025,1,1)
d2 = date(2026,2,10)
print((d2 - d1).days)
```

```python
dob = date(2005,2,17)
today = date.today()
if today.month == dob.month and today.day == dob.day:
    print("Happy Birthday!")
else:
    print("Today is not your birthday.")
```
